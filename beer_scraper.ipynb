```python
# get all links for top 250 beers
top250 = session.get('https://www.beeradvocate.com/beer/top-rated/')
soup = BeautifulSoup(top250.content, 'html5lib')
dom = etree.HTML(str(soup))
beers = {}
for i in range(2, 252):
  link = dom.xpath(f'//*[@id="ba-content"]/table/tbody/tr[{i}]/td[2]/a')
  beer_name = dom.xpath(f'//*[@id="ba-content"]/table/tbody/tr[{i}]/td[2]/a/b')
  beer_type = dom.xpath(f'//*[@id="ba-content"]/table/tbody/tr[{i}]/td[2]/span/a[2]')
  beers[beer_name[0].text] = {'type': beer_type[0].text, 'link': link[0].get('href')}
print(beers)
```

```python
import pandas as pd
import numpy as np

beer_df = pd.DataFrame(columns=['name', 'review', 'rating', 'type'])
beer_df

base_url = 'https://www.beeradvocate.com'
for beer in beers:
  beer_name = beer
  beer_type = beers[beer]['type']
  num = 0
  while num < 140:
    try:
      url = base_url + beers[beer]['link'] + '?view=beer&show=recent&order=high&start=' + str(num) + '#lists'
      beer_page = session.get(url)
      soup = BeautifulSoup(beer_page.content, 'html5lib')
      comments = soup.findAll('div', attrs={'class': 'user-comment'})
      for i, comment in enumerate(comments):
        dom = etree.HTML(str(comment))
        review = dom.xpath('//*[@id="rating_fullview_content_2"]/div')
        rating = comment.find('span', attrs={'class': 'BAscore_norm'})
        if review:
          beer_review = review[0].text
          beer_rating = rating.text
          beer_df.loc[len(beer_df)] = [beer_name, beer_review, beer_rating, beer_type]
      num += 20
    except:
      break
```
